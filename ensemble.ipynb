```python
import pandas as pd
import numpy as np
import os
```

```python
#Avg
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Define the path to the records folder
records_path = "./records"

# List to store DataFrames
dataframes = []

# Read all CSVs in the folder and combine them
for file in os.listdir(records_path):
    if file.endswith(".csv"):
        if "Test" in file:
            filepath = os.path.join(records_path, file)
            df = pd.read_csv(filepath)
            print(filepath)
            dataframes.append(df)

# Concatenate all dataframes
combined_df = pd.concat(dataframes, ignore_index=True)

# Ensure columns 'y_true' and 'y_predict' exist
if 'y_true' in combined_df.columns and 'y_predict' in combined_df.columns:
    # Group by y_true and average y_predict
    averaged_predictions = combined_df.groupby('y_true')['y_predict'].mean().reset_index()
    y_true = averaged_predictions['y_true']
    y_predict_avg = averaged_predictions['y_predict']
    
    # Calculate MAE and RMSE
    mae = mean_absolute_error(y_true, y_predict_avg)
    rmse = np.sqrt(mean_squared_error(y_true, y_predict_avg))
    
    # Print results
    print(f"Mean Absolute Error (MAE): {mae}")
    print(f"Root Mean Squared Error (RMSE): {rmse}")
else:
    print("The required columns ('y_true', 'y_predict') are not present in the CSV files.")
```
