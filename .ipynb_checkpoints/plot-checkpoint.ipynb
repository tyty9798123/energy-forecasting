```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 設置CSV檔案所在的資料夾路徑
folder_path = 'records/TCN_FINAL/C_tcn_2_32_1'

# 讀取資料夾中所有CSV檔案
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# 合併所有CSV檔案
combined_df = pd.DataFrame()
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    combined_df = pd.concat([combined_df, df])

# 重置索引
combined_df.reset_index(drop=True, inplace=True)

# 自動獲取欄位名稱
columns = combined_df.columns
x_column = columns[0]
y_column = columns[1]

start=1800
end=1850
combined_df = combined_df[start:end]

# 使用獲取到的欄位名稱繪圖
plt.figure(figsize=(10, 6))
plt.plot(combined_df[x_column], label=x_column)
plt.plot(combined_df[y_column], label=y_column)

plt.title('Combined Data Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

def calculate_rmse_mae(y_true, y_pred):
    """
    計算並輸出RMSE和MAE。

    參數:
    y_true (array-like): 真實值
    y_pred (array-like): 預測值

    返回:
    tuple: 包含RMSE和MAE的元組
    """
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    
    print("RMSE:", rmse)
    print("MAE:", mae)

    return rmse, mae
```
